# *RULE OF LAW -HUKUKUN ÜSTÜNLÜĞÜ*

**Hukukun üstünlüğü, bireylerin hak ve özgürlüklerinin korunmasında temel bir öneme sahiptir. Bu ilke, devletin güçlerini sınırlandırarak, bireylerin yasalar önünde eşit muamele görmesini sağlar ve keyfi yönetim uygulamalarına karşı bir güvence oluşturur. Hukukun üstünlüğü, insanların insana saygı ve adalet arayışlarında güvenilir bir zemin sunar; böylece bireyler, kendilerini güvende hissettiğinde özgürlüklerini daha etkin bir şekilde kullanabilirler. Bir ülkede hukukun üstünlüğünün yüksek olması, bireylerin yönetime olan güvenini artırırken, sosyal uyumu ve ekonomik kalkınmayı da destekler. Bu durum, toplumsal barışı pekiştirir ve bireylerin bireysel potansiyellerini gerçekleştirmelerine olanak tanır. Dolayısıyla, hukukun üstünlüğü, hem bireylerin özgürlük ve haklarının korunmasında hem de toplumların gelişiminde kritik bir role sahiptir.**

```python
rol = pd.DataFrame( data=data, columns = ['year','countries','region','pf_rol_procedural', 'pf_rol_civil', 'pf_rol_criminal','pf_rol'])
rol
```

```python
rol.info()
```

```python
rol.isna().sum()
```

**Şimdi Hukukun üstünlüğünü analiz edicem fakat veri setimde eksik veriler çıktı. Bu eksik veriler analizimde sorun yaratacağı için eksik verilere yaklaşım stratejileri uygulayacağım.
Eksik Veri Yöntemleri**
* Silmek
* Bir önceki satırla doldurmak
* Ortalama mod medyan gibi değerler ile doldurmak

**Ben verimde çok fazla satır bulunduğu için silmeyi tercih ediyorum.**

```python
rol1 = rol.dropna()
rol1
```

```python
rol1.isna().sum() 
```

**Görüldüğü üzere eksik verilerden kurtulmuş olduk şimdi analizimize başlayabiliriz.**

```python
sns.barplot(rol1, x="pf_rol", y="region")
```

***Bölgelere göre hukukun üstünlüğü puan grafiğidir. Görüldüğü üzere yukarıdaki grafikte hukukun üstünlüğü en yüksek olan bölgeler Batı Avrupa ve Okyanusya ülkeleri denilebilir. En düşük ise Sahra Altı Bölgesi Ülkeleri diyebiliriz.***

**Sadece Türkiye ile Batı Avrupa bölgesinde ki bir ülkenin hukuka karşı ölçütlerinin ortalamasının oranlarını karşılaştırmak istiyorum.**

```python
m = rol1["countries"].isin(["Turkey"]) 
rol1[m] 
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(15,5))
sns.barplot(x = "year", y= "pf_rol" ,data = rol1[m] )
```

**Bu grafiğe göre anlaşılıyor ki yıllara göre Türkiyenin Dünya İnsan Özgürlüğü indeksine göre Hukuk'un üstünlüğü skoru düşmüştür.**

```python
ort = rol1.groupby("region")[['pf_rol_procedural', 'pf_rol_civil', 'pf_rol_criminal']].agg(["mean"]).reset_index()
ort
```

**Bölgelere göre hukukun prosudur skoru, sivil ve suça karşı uygulanabilirliğinin ortalamalarını görüyoruz.**

```python
ort.max() 
```

```python
ort.min()
```

**Ortalamala göre en yüksek sonuçlar Batı Avrupa iken en düşük ortalamalara sahip bölge Orta Asya ülkeleri diyebiliriz.**

# *SECURITY AND SAFETY - EMNİYET VE GÜVENLİK*

**Emniyet ve güvenlik, toplumların düzenini sağlamak, bireylerin can ve mal güvenliğini korumak için temel unsurlardır. Emniyet, genellikle suçların önlenmesi, toplum düzeninin sağlanması ve halkın güvenliğini tehlikeye atan durumların engellenmesiyle ilişkilidir. Güvenlik ise, hem bireylerin fiziksel güvenliğini hem de toplumun huzurunu koruma amacını taşır.**

```python
sec = pd.DataFrame( data=data, columns = ['year','countries','region','pf_ss_homicide','pf_ss_disappearances_violent'])
sec
```

```python
sec.info()
```

```python
sec.isnull().sum()
```

**Emniyet ve Güvenlik kısmında da eksik veriler bulunduğunu fark ettim daha önce yaptığım gibi verimde çok fazla satır bulunduğu için silmeyi tercih ediyorum.**

```python
sec1 = sec.dropna()
sec1
```

```python
sec1.isnull().sum()
```

**Görüldüğü üzere eksik verilerden kurtulduk. Analizimize devam edelim.**

```python
sns.barplot(data=sec1, x="pf_ss_homicide", y="region")
```

***Bölgelere göre emniyet ve güvenliğin grafiğidir. Görüldüğü üzere yukarıdaki grafikte emniyet ve güvenliğin en yüksek olan bölgeler Batı Avrupa ülkeleri denilebilir. En düşük ise Latin Amerika ve Sahra Altı Afrika görece düşük skora sahip diyebiliriz.***

**Şimdi de sadece Türkiye ile Batı Avrupa bölgesinde ki bir ülkenin emniyet ve güvenliğe karşı ölçütlerinin ortalamasının oranlarını karşılaştırmak istiyorum.** 

```python
trsec = sec1["countries"].isin(["Turkey"]) 
sec1[trsec] 
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(15,5))
sns.barplot(x = "year", y= "pf_ss_homicide" ,data = sec1[trsec] )
```

**Bu grafiğe göre yıllara göre Türkiyenin emniyet ve güvenliği zaman zaman azalsade son yıllarda artışa geçtiği gözlemlenmiştir.**

```python
ortsec = sec1.groupby("region")[['pf_ss_homicide','pf_ss_disappearances_violent']].agg(["mean"]).reset_index()
ortsec
```

**Bölgelere göre emniyet ve güvenlik skoru, kaybolmalar ve şiddete karşı uygulanabilirliğinin ortalamalarını görüyoruz.**

```python
ortsec.max()
```

```python
ortsec.min()
```

**Ortalamala göre en yüksek sonuçlar Batı Avrupa iken en düşük ortalamalara sahip bölge Orta Asya ülkeleri diyebiliriz.**
